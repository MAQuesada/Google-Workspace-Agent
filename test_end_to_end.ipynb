## Chat Simulation

```python
import asyncio
import ipywidgets as widgets
from IPython.display import display, Markdown


# Input widget
input_box = widgets.Text(
    value='',
    placeholder='Type your question and press Enter...',
    description='User:',
    layout=widgets.Layout(width='100%')
)

# Output widget
output_box = widgets.Output()

# Async handler
async def on_submit_enter(_):
    question = input_box.value
    input_box.value = ''  # clean the input box

    # Process the question
    response = await orchestrator_graph.ainvoke(
        {"user_input": question, "user_id": user_id}, config)
    answer = response["messages"][-1].content

    output_box.append_display_data(Markdown(f"**🧑 User:** {question}"))
    output_box.append_display_data(Markdown(f"**🤖 Bot:**\n {answer}"))
    output_box.append_display_data(Markdown("---"))

input_box.on_submit(lambda _: asyncio.ensure_future(on_submit_enter(_)))

display(input_box)
display(output_box)
# create a draft for my contact carlos thanking him for his efforts this week in the sales department. also include my free time for tomorrow from 9am to 5pm to set up a meeting with him. Please use my university account.
```
