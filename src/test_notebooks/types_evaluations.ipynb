# Evaluations in Agent Systems

## Ways to test/evaluate an agent system

| Evaluation Method (Description) | Task Performance | System Performance | Security & Robustness | Ethics & Alignment |
|-----------------------|------------------|--------------------|-----------------------|--------------------|
| **LLM-as-a-Judge** – Uses a stronger or separate LLM to score, compare, or critique outputs for relevance, correctness, tone, completeness, or reasoning quality. Fast, scalable, but may inherit model biases. | ✅ | ❌ | ⚠️ (limited) | ⚠️ (surface-level) |
| **Rule-Based Testing** – Classic software testing adapted to agents, using assertions/validation checks like schemas, regex, or tool usage logs to ensure rules are followed. Great for structured output. | ✅ | ✅ | ⚠️ (for expected paths) | ❌ |
| **Standardized Benchmarks** – Public datasets/test suites for common tasks or safety measures (e.g., HumanEval, TruthfulQA). Community-aligned and reproducible, but risk of overfitting. | ✅ | ❌ | ⚠️ (if benchmark includes) | ⚠️ (if benchmark includes) |
| **Simulated Experiments** – Controlled scenarios to test agent behavior, planning, tool use, and adaptability. Can stress-test under load or varied inputs; must avoid excessive scenario explosion. | ✅ | ✅ | ✅ | ⚠️ (depends on design) |
| **Golden Datasets** – Curated, trusted examples with correct outputs for direct accuracy measurement in low-ambiguity tasks. Reliable but costly to build and not suited for subjective qualities. | ✅ | ❌ | ❌ | ❌ |
| **Human Review** – Experts review outputs based on rubrics or judgment, ideal for subjective or high-stakes tasks. High-quality but slow, costly, and prone to reviewer bias. | ✅ | ❌ | ⚠️ (behavioral nuance) | ✅ |
| **Red Teaming** – Adversarial testing with malicious, tricky, or confusing inputs to find system weaknesses and stress-test edge cases. Manual but critical for robustness and alignment. | ⚠️ (failure cases) | ❌ | ❌ | ✅ |


## The System Design Dimensions That Shape What You Should Measure

Not all agentic systems are built the same — and not all should be evaluated the same way.

To choose the right metrics, you first need to understand what kind of system you’re building. That means asking:

- What does the system produce?
- Who is it for?
- How complex is it under the hood?
- And how do users expect to interact with it?

To make that easier, we’ve distilled these questions into seven key dimensions — a compact framework that helps you identify what matters most in your use case.
| System Design Dimension | What It Is | How It Impacts Metric Selection | Metrics & Evaluation Approach |
|-------------------------|------------|----------------------------------|--------------------------------|
| **User Type** | Who the system is for: internal teams or external users | Affects tolerance for errors, polish, and evaluation strictness | **Internal:** focus on task success, intermediate step accuracy, functional reliability. Errors acceptable if easy to debug or recover from. **External:** include tone, consistency, presentation, and trust metrics (e.g., user satisfaction, human/LLM reviews). |
| **System Stakes** | Whether the system operates in high-risk contexts (e.g., healthcare) | Raises bar for safety, robustness, and fallback evaluations | **Low stakes:** prioritize usefulness and relevance; creativity metrics acceptable. **High stakes:** include safety, robustness under stress, critical result validation, fallback coverage, and absence of catastrophic failures. |
| **Goal Clarity** | How well-defined the task or objective is | Determines whether you can use standard metrics or need custom, rubric-based ones | **High clarity:** use objective metrics like exact match, precision, recall, F1, task success. **Low clarity:** apply rubrics, human/LLM judgment, reasoning heuristics, and subjective quality ratings. |
| **Final Outcome Type** | What the system is expected to produce | Determines whether to use generation or task-completion metrics | **Generation:** evaluate coherence, fluency, relevance, tone (LLM-as-a-Judge, human review). **Action:** measure task success rate, tool/API usage accuracy, and side-effect safety. |
| **Interaction Mode** | Whether the user expects a one-shot or multi-step experience | Determines whether intermediate step correctness matters | **One-shot:** assess first-response accuracy, completeness, and clarity. **Multi-step:** measure intermediate step correctness, memory/context persistence, cross-turn consistency, and convergence speed. |
| **Grounding Needs** | Whether output must stay tied to source data | Determines importance of factuality and faithfulness metrics | **With grounding:** measure source faithfulness, textual similarity, and hallucination rate. **Without grounding:** focus on usefulness, originality, and creativity. |
| **Goal Complexity** | How challenging the task is to complete reliably | Drives need for intermediate and chain-level metrics | **Low complexity:** evaluate final output only. **High complexity:** measure planning quality, reasoning, tool accuracy, intermediate success rate, and chain consistency. |


## Example: Evaluating the Agentic Authoring Assistant

The **Agentic Authoring Assistant (A3)** on the Ready Tensor platform helps users auto-generate publication metadata — including the title, tags, TL;DR, and references. Each output improves how projects are presented, discovered, and shared.

We will evaluate each of the four outputs from the system, and then also assess the overall coherence of the system’s outputs. This will help us understand how well A3 meets user needs, maintains quality, and delivers a consistent experience.


### Dimensional Breakdown

To evaluate this system, we apply the **7 design dimensions** to each component and derive what matters most — and how to measure it.

| Component       | User Type | System Stakes | Goal Clarity | Outcome Type | Interaction Mode | Grounding Needs | Goal Complexity |
|-----------------|-----------|---------------|--------------|--------------|------------------|-----------------|-----------------|
| **Title**       | External  | Low           | High         | Generation   | One-shot         | Yes             | Low             |
| **Tags**        | External  | Low           | High         | Generation   | One-shot         | Yes             | Low             |
| **TL;DR Summary** | External  | Low           | High         | Generation   | One-shot         | Yes             | Low             |
| **References**  | External  | Low           | High         | Generation   | One-shot         | Yes             | Medium          |
| **Coherence (All)** | External  | Low           | High         | Generation   | One-shot         | Yes             | Medium          |

---


#### **User Type**
All components are designed for external users. While the platform hosts their work, A3 is helping them publish and share it with others — so quality and presentation matter.

**Implication:** Evaluation must include **polish, tone, and presentation quality** — not just task correctness.


#### **System Stakes**
We marked these as low stakes, since A3 is an assistant, not a decision-making agent. Users can always ignore suggestions or revise them.  
If A3 were a paid product or embedded into a professional workflow, stakes might rise to medium due to higher expectations.

**Implication:** Evaluation can tolerate **minor errors**, but should still surface inconsistencies or issues that affect **user trust**.


#### **Goal Clarity**
All tasks have a clear objective — generate a title, a summary, some tags, etc. There’s little ambiguity about what a "complete" result looks like.

**Implication:** **Reference-based and automatic metrics** (e.g., similarity scores, exact match, relevance ratings) are appropriate and reliable.


#### **Outcome Type**
Every component produces **generated text**.

**Implication:** Use metrics suited to **generation** — such as **fluency, clarity, faithfulness, and relevance** — rather than task completion or action success.

#### **Interaction Mode**
A3 is single-turn — each feature runs once per user click. There’s no back-and-forth interaction or refinement loop (yet).

**Implication:** Focus on **immediate output quality**; no need for session-level or memory persistence checks.


#### **Grounding Needs**
All outputs must be grounded in the **user’s publication**. Hallucinations or off-topic suggestions are a problem.

**Implication:** **Faithfulness scoring is essential** — using LLM judgment, textual similarity, or rule-based grounding checks.


#### **Goal Complexity**
Most components are **low complexity**. References require **external search, filtering, and justification**.  
Coherence across outputs is **medium complexity** due to subjectivity.

**Implication:** Use **simple outcome metrics** for low-complexity tasks, and **structured rubrics with LLM-as-a-Judge** for subjective ones.


### Evaluation Approach

Each part of the A3 system is evaluated **based on final outputs**, not internal reasoning or multi-step coordination.  
Since tasks are simple, well-scoped, and one-shot, the focus is on **output quality**.

We use a mix of:
- **LLM-as-a-Judge**
- **Golden datasets**
- **Rule-based methods**

Our **golden dataset** comes from high-quality publications already on Ready Tensor, curated by experts.

| Component       | Focus Area                      | Evaluation Methods                | Metrics                                               |
|-----------------|----------------------------------|------------------------------------|-------------------------------------------------------|
| **Title**       | Relevance, clarity, engagement   | LLM-as-a-Judge, Golden dataset     | Relevance rating, Similarity score, Faithfulness score |
| **Tags**        | Accuracy, topical coverage       | LLM-as-a-Judge, Golden dataset     | Relevance rating, Jaccard similarity                  |
| **TL;DR Summary** | Faithfulness, clarity, brevity  | LLM-as-a-Judge, Golden dataset     | Relevance rating, Similarity score, Faithfulness score |

---

**Note:**  
Because all components are single-turn generations, we don’t need metrics for:
- Tool routing  
- Agent collaboration  
- Memory consistency  

In more complex systems, evaluations might happen at **step level** or across **decision chains**.  
Here, **output quality** is the main focus, with methods chosen to keep judgments **reliable and scalable**.


##  **Key Metrics for Evaluating Agents & LLM Systems**

* **Accuracy**:  Measures whether the generated answer exactly matches the correct expected answer. Best for closed tasks with one right answer.


* **Faithfulness**
  Checks if the response is factually correct based on the given evidence or sources, without making things up (*avoids hallucinations*).

* **Relevance**
  Assesses whether the response directly addresses the question or context without drifting into unrelated content.

* **Semantic Similarity**
  Compares the meaning of the response with the reference answer, even if the wording is different. Often measured using embeddings or cosine similarity.

* **Jaccard Similarity**
  Measures overlap between two sets (e.g., tags, references):

  $$
  \frac{\text{Shared elements}}{\text{Total unique elements}}
  $$

* **Coherence**
  Evaluates if the response flows logically, with no contradictions between its parts.

* **Toxicity**
  Detects offensive, discriminatory, or harmful language.

* **Bias Detection**
  Identifies whether the response shows unwanted prejudice or unfair favoritism.

* **Coverage**
  Measures how completely the answer covers all required key points.

* **Conciseness**
  Checks whether the answer communicates the needed information without unnecessary length or repetition.
