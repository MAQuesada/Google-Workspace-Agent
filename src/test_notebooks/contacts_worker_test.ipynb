# Contacts Worker Test

```python
from dotenv import load_dotenv
import os
from IPython.display import Markdown
from langchain_core.messages import HumanMessage


import sys
from pathlib import Path

project_root = Path(".").resolve().parent.parent
sys.path.append(str(project_root))
os.chdir(project_root)


load_dotenv()

test_user_email = "manuel.quesada@student.uhasselt.be"
user_id = "manuel-test"
config = {"configurable": {"thread_id": "testing_state_graph01"}}
```

```python
from IPython.display import Markdown
from langchain_core.messages import HumanMessage
from agents.contacts.worker import google_contact_worker
```

## To find by mail or name

```python
response = await google_contact_worker.ainvoke(
    {"workers_messages": [HumanMessage(content="Tell me all my contacts related to 'alejandro'")], "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```

## Create and update

```python
response = await google_contact_worker.ainvoke(
    {"workers_messages": [HumanMessage(content="Create a contact named Juancito")], "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```

```python
response = await google_contact_worker.ainvoke(
    {"workers_messages": [HumanMessage(content="change the name of the contact Juancito to Jesus and add the phone number 545011001-111")], "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```

```python
response = await google_contact_worker.ainvoke(
    {"workers_messages": [HumanMessage(content="delete contact Juan")], "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```

```python
response = await google_contact_worker.ainvoke(
    {"workers_messages": [HumanMessage(content="ok, delete Juan@mail.com")], "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```

```python
response = await google_contact_worker.ainvoke(
    {"workers_messages": [HumanMessage(content="tell me all my contacts related to juan")], "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```

```python
response = await google_contact_worker.ainvoke(
    {"workers_messages": [HumanMessage(content=" create the contact Juanito(uncle) with phone 54594034")], "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```

Super cool!!! Apesar de que en el prompt se le esta indicando que debe usar una sola tool; en este caso el LLM fue capaz de buscar los contactos que incluian a 'juanito' y mandarlos a borrar uno a uno. En situaciones como esta es donde se nota la diferencia de un workflow tradicional y un AI AGENT.
