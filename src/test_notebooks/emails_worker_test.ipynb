## Setup

```python
from dotenv import load_dotenv
import os
from IPython.display import Markdown
from langchain_core.messages import HumanMessage


import sys
from pathlib import Path

project_root = Path(".").resolve().parent.parent
sys.path.append(str(project_root))
os.chdir(project_root)


load_dotenv()
os.environ["TESTING"] = "true"

test_user_email = "manuel.quesada@student.uhasselt.be"
user_id = "manuel-test"
config = {"configurable": {"thread_id": "testing_state_graph01"}}
```

```python
from IPython.display import Markdown
from langchain_core.messages import HumanMessage
from agents.emails.worker import google_email_worker
```

### Available Tools

```python
response = await google_email_worker.ainvoke(
    {"workers_messages": [HumanMessage(content="hello. How you can assist me?")],
     "user_id": user_id,
     "account_name": test_user_email
     }, config)
Markdown(response["workers_messages"][-1].content)
```
