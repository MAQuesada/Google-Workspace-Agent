````python
response = await orchestrator_graph.ainvoke(
    {
        "user_input": "create a draft for my contact carlos thanking him for his efforts this week in the sales department. also include my free time for tomorrow from 9am to 5pm to set up a meeting with him. Please use my university account.",
        "user_id": user_id,
    },
    config,
)
Markdown(response["messages"][-1].content)
````

In this case: the orchestrator plannig the following actions:

```
{"reasoning":"The user's request involves creating a draft email for a contact named Carlos and includes checking the user's free time for tomorrow. Therefore, we need to use the 'contacts_manage' to find Carlos's email address and 'calendar_manage' to check the user's availability. Since the user specified the university account, we will use this account for both managers. First, we will use 'date_manage' to determine the specific date for 'tomorrow' and then check the calendar for free time slots. Finally, we will create the draft email using 'email_manage'.","managers":[{"route_manager":"date_manage","query":"Determine the date for 'tomorrow' from the current date."},{"route_manager":"contacts_manage","query":"Find the contact information for Carlos in the university account (manuel.quesada@uhasselt.be)."},{"route_manager":"calendar_manage","query":"Check for free time slots on the determined date from 9am to 5pm in the university account (manuel.quesada@uhasselt.be)."},{"route_manager":"email_manage","query":"Create a draft email in the university account (manuel.quesada@uhasselt.be) to Carlos, thanking him for his efforts this week in the sales department. Include the user's free time for tomorrow from 9am to 5pm to set up a meeting. Best regards, Manuel-test."}]}
```

But as the carlos contact doesn't have an email address, the flow was stopped and the user was asked to provide the email address.

````python
response = await orchestrator_graph.ainvoke(
    {
        "user_input": "you are rigth, use the email calos@mail.com and please update the contact with this email",
        "user_id": user_id,
    },
    config,
)
Markdown(response["messages"][-1].content)
````
